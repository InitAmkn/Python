```python
import sympy
import matplotlib
from sympy.plotting import plot
```

```python
x = sympy.Symbol('x')
print(x)
```

```python
f_x = -12*x**4*sympy.sin(sympy.cos(x)) - 18*x**3+5*x**2 + 10*x - 30
print(f_x)
```

```python
#print(sympy.solve(f_x,x))
```

```python
pl = plot(f_x, ylim=(-10000, 10000))
```

```python
#поиск корней
def find_x(f_x, x_start, x_finish, accuracy):
    i = x_start
    min = f_x.subs(x,i)
    if min<0:
            min *= -1
    return_x = 0
    while i<=x_finish:
        temp = f_x.subs(x,i)
        if temp<0:
            temp *= -1
        i += accuracy
        if temp<min:
            return_x = i
            min = temp
    return return_x
    
   
    
print(find_x(f_x,-7.7,-7.6, 0.0001))
print(find_x(f_x,-5.3,-5, 0.0001))
print(find_x(f_x,-1.5,-1, 0.0001))
print(find_x(f_x,2,2.5, 0.0001))
print(find_x(f_x,4.3,4.5, 0.0001))
print(find_x(f_x,7.8,8.5, 0.0001))
```

```python
def find_x(f_x, x_start, x_finish, accuracy):
    i = x_start
    max = f_x.subs(x,i)
    if max<0:
            max *= -1
    return_x = 0
    while i<=x_finish:
        temp = f_x.subs(x,i)
        if temp<0:
            temp *= -1
        i += accuracy
        if temp>max:
            return_x = i
            max = temp
    return (return_x, max)

print(find_x(f_x,0,2, 0.0001))
```

```python
diff_x = f_x.diff()
pl = plot(diff_x, ylim=(-1000, 1000))
```

```python
print(find_x(diff_x, -10.1, -9.8, 0.0001))
print(find_x(diff_x, -7.5, -6, 0.0001))
print(find_x(diff_x, -4.8, -3, 0.0001))
print(find_x(diff_x, -1, 0, 0.0001))
print(find_x(diff_x, 0, 1, 0.0001))
print(find_x(diff_x, 1, 2.4, 0.0001))
print(find_x(diff_x, 2.8, 4.5, 0.0001))
print(find_x(diff_x, 5.5, 7.5, 0.0001))
print(find_x(diff_x, 8, 10, 0.0001))
```
